```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/drive/MyDrive/파이널 3조/워드클라우드/리뷰_AI행삭제.csv', encoding='utf-8')
```

```python
# 폰트 경로 지정
font_path = '/content/drive/MyDrive/파이널 3조/워드클라우드/NanumBarunpen.ttf'
```

```python
# 워드클라우드를 생성할 텍스트 데이터를 추출
text_data = ' '.join(data['리뷰내용'])

# WordCloud 객체를 생성하고 워드클라우드를 생성
wordcloud = WordCloud(font_path=font_path, background_color='white').generate(text_data)

# 워드클라우드 시각화
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize
from collections import Counter
```

```python
words = word_tokenize(text_data)

# 단어별 빈도수 계산
word_counts = Counter(words)

# WordCloud 객체를 생성하고 워드클라우드를 생성
wordcloud = WordCloud(font_path=font_path, background_color='white').generate_from_frequencies(word_counts)

# 워드클라우드 시각화
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# 워드클라우드에서 나온 단어 목록 가져오기
cloud_words = list(filtered_word_counts_g.keys())

# 해당 단어가 포함된 리뷰 추출
matching_reviews = [review for review in Glossy_reviews if any(word in review for word in cloud_words)]

# 추출된 리뷰 출력
for review in matching_reviews:
    print(review)
```
